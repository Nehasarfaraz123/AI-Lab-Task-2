```python
# Simple Hill Climbing Algorithm in Python
# Hill Climbing is a search algorithm.
# It moves step by step toward a better solution.
# Here, we try to maximize the function:
#
#           f(x) = -(x^2) + 5
#
# The highest value of this function is at x = 0

# Objective Function
# This function calculates the value of f(x)

def objective_function(x):
    return -(x ** 2) + 5

# Hill Climbing Function
def hill_climbing(start, step_size, max_iterations):

    # Current starting position
    current = start

    # Calculate value of current position
    current_value = objective_function(current)

    # Display initial state
    print("Starting Point:", current)
    print("Starting Value:", current_value)
    print("--------------------------------")

    # Repeat until maximum iterations
    for i in range(max_iterations):

        # Generate neighboring states
        left_neighbor = current - step_size
        right_neighbor = current + step_size

        # Calculate values of neighbors
        left_value = objective_function(left_neighbor)
        right_value = objective_function(right_neighbor)


        # Compare neighbors with current state
        # Move toward better solution

        # If left neighbor is better
        if left_value > current_value:

            current = left_neighbor
            current_value = left_value

        # If right neighbor is better
        elif right_value > current_value:

            current = right_neighbor
            current_value = right_value

        # If no better neighbor exists
        else:
            print("No better neighbor found.")
            print("Algorithm stopped.")
            break

        # Display current progress
        print("Iteration:", i + 1)
        print("Current State:", current)
        print("Current Value:", current_value)
        print("--------------------------------")

    # Return best solution found
    return current, current_value

# Run the Hill Climbing Algorithm
best_state, best_value = hill_climbing(
    start=4,            # Starting position
    step_size=1,        # Step movement size
    max_iterations=20   # Maximum loop iterations
)

# Final Output
print("\nBest Solution Found:", best_state)
print("Best Value:", best_value)
```

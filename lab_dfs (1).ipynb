```python
# Using a Python dictionary to act as an adjacency list
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = set()  # Set to keep track of visited nodes

def dfs(visited, graph, node):  # function for DFS
    if node not in visited:
        print(node, end=" ")
        visited.add(node)

        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
print("Following is the Depth-First Search")
dfs(visited, graph, '5')
```

```python
from collections import defaultdict

# This class represents a directed graph
class Graph:

    # Constructor
    def __init__(self):
        self.graph = defaultdict(list)

    # function to add an edge to graph
    def addEdge(self, u, v):
        self.graph[u].append(v)

    # DFS helper function
    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")

        # Recur for all adjacent vertices
        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)

    # DFS traversal function
    def DFS(self):
        visited = [False] * (max(self.graph) + 1)

        for i in range(len(visited)):
            if not visited[i]:
                self.DFSUtil(i, visited)


# Driver code
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal")
g.DFS()
```

```python
# DFS for Travelling Salesman Problem (Arad to Bucharest)

graph = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu': [('Bucharest', 90)],
    'Urziceni': [('Bucharest', 85), ('Hirsova', 98), ('Vaslui', 142)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Eforie': [('Hirsova', 86)],
    'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
    'Iasi': [('Vaslui', 92), ('Neamt', 87)],
    'Neamt': [('Iasi', 87)]
}

def dfs(node, goal, visited, path, cost):
    visited.add(node)
    path.append(node)

    if node == goal:
        return path, cost

    for neighbor, weight in graph[node]:
        if neighbor not in visited:
            result = dfs(neighbor, goal, visited, path.copy(), cost + weight)
            if result:
                return result

    return None


# Driver Code
start = 'Arad'
goal = 'Bucharest'

result = dfs(start, goal, set(), [], 0)

if result:
    path, cost = result
    print("DFS Path from Arad to Bucharest:")
    print(" -> ".join(path))
    print("Total Distance:", cost)
else:
    print("No path found")
```

```python
from collections import deque

# Graph
graph = {
    'Arad': ['Zerind', 'Sibiu', 'Timisoara'],
    'Zerind': ['Arad', 'Oradea'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Rimnicu Vilcea': ['Sibiu', 'Craiova', 'Pitesti'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Pitesti': ['Rimnicu Vilcea', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Hirsova', 'Vaslui'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Eforie': ['Hirsova'],
    'Vaslui': ['Urziceni', 'Iasi'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Neamt': ['Iasi']
}

# -------- BFS --------
def bfs(start, goal):
    visited = set()
    queue = deque([start])
    count = 0

    while queue:
        node = queue.popleft()

        if node not in visited:
            print("BFS Visiting:", node)
            visited.add(node)
            count += 1

            if node == goal:
                break

            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    print("BFS Nodes Visited:", count)
    print("--------------------------------")


# -------- DFS --------
def dfs(node, goal, visited):
    print("DFS Visiting:", node)
    visited.add(node)

    if node == goal:
        return True, 1

    count = 1

    for neighbor in graph[node]:
        if neighbor not in visited:
            found, sub_count = dfs(neighbor, goal, visited)
            count += sub_count
            if found:
                return True, count

    return False, count


# =========================
# TESTING
# =========================

start = 'Arad'

# ✅ CASE 1 (BFS better)
goal = 'Sibiu'

# ❗ CASE 2 (DFS better)
# goal = 'Neamt'

print("START:", start)
print("GOAL:", goal)

print("\n===== BFS =====")
bfs(start, goal)

print("\n===== DFS =====")
found, dfs_count = dfs(start, goal, set())
print("DFS Nodes Visited:", dfs_count)
```

Activity 1 — Four Cyan Squares

```python
from pygame.locals import *

class Square(pygame.sprite.Sprite):
    def __init__(self):
        super(Square, self).__init__()
        self.surf = pygame.Surface((25, 25))
        self.surf.fill((0, 200, 255))
        self.rect = self.surf.get_rect()

screen = pygame.display.set_mode((800, 600))

square1 = Square()
square2 = Square()
square3 = Square()
square4 = Square()

screen.fill((0, 0, 0))
screen.blit(square1.surf, (40, 40))
screen.blit(square2.surf, (40, 530))
screen.blit(square3.surf, (730, 40))
screen.blit(square4.surf, (730, 530))

pygame.display.flip()
pygame.image.save(screen, "activity1.png")
print("✅ Done!")
```

Lab Task 1 — Basic Snake Game

```python
import pygame
import random
from IPython.display import Image as IPImage, display

screen = pygame.display.set_mode((600, 400))

# Colors
BLACK  = (0,   0,   0)
WHITE  = (255, 255, 255)
GREEN  = (0,   200, 0)
DGREEN = (0,   150, 0)
RED    = (213, 50,  80)

WIDTH, HEIGHT = 600, 400
BLOCK = 20

font       = pygame.font.SysFont("arial", 24, bold=True)
small_font = pygame.font.SysFont("arial", 18)

# ── Functions ──────────────────────────────────────────
def draw_snake(surface, snake_list):
    for i, block in enumerate(snake_list):
        color = DGREEN if i == 0 else GREEN
        pygame.draw.rect(surface, color,
                        [block[0], block[1], BLOCK-2, BLOCK-2], border_radius=4)

def draw_food(surface, food):
    pygame.draw.circle(surface, RED,
                      (food[0] + BLOCK//2, food[1] + BLOCK//2), BLOCK//2)

def draw_score(surface, score):
    text = font.render(f"Score: {score}", True, WHITE)
    surface.blit(text, (8, 8))

def draw_gameover(surface, score):
    overlay = pygame.Surface((WIDTH, HEIGHT))
    overlay.set_alpha(180)
    overlay.fill(BLACK)
    surface.blit(overlay, (0, 0))
    go_text = font.render("GAME OVER", True, RED)
    sc_text = small_font.render(f"Final Score: {score}", True, WHITE)
    surface.blit(go_text, (WIDTH//2 - go_text.get_width()//2, HEIGHT//2 - 40))
    surface.blit(sc_text, (WIDTH//2 - sc_text.get_width()//2, HEIGHT//2 + 10))

# ── Game State ─────────────────────────────────────────
snake     = [[200, 200], [180, 200], [160, 200]]
direction = [BLOCK, 0]
food      = [random.randrange(0, WIDTH//BLOCK)  * BLOCK,
             random.randrange(0, HEIGHT//BLOCK) * BLOCK]
score     = 0
game_over = False

# ── Simulate Game (auto moves for Colab) ───────────────
moves = (
    [BLOCK,  0] * 80  +   # right
    [0,  BLOCK] * 60  +   # down
    [-BLOCK, 0] * 80  +   # left
    [0, -BLOCK] * 60      # up
)

screenshots = []

for frame, move in enumerate(moves):
    if game_over:
        break

    direction = list(move) if isinstance(move, tuple) else direction

    # Apply moves in chunks
    if   frame < 80:  direction = [BLOCK,  0]
    elif frame < 140: direction = [0,  BLOCK]
    elif frame < 220: direction = [-BLOCK, 0]
    elif frame < 280: direction = [0, -BLOCK]

    new_head = [snake[0][0] + direction[0],
                snake[0][1] + direction[1]]
    snake.insert(0, new_head)

    # Food check
    if snake[0][0] == food[0] and snake[0][1] == food[1]:
        score += 10
        food = [random.randrange(0, WIDTH//BLOCK) * BLOCK,
                random.randrange(0, HEIGHT//BLOCK) * BLOCK]
    else:
        snake.pop()

    # Wall collision
    if (snake[0][0] < 0 or snake[0][0] >= WIDTH or
        snake[0][1] < 0 or snake[0][1] >= HEIGHT):
        game_over = True

    # Self collision
    if snake[0] in snake[1:]:
        game_over = True

    # Draw
    screen.fill(BLACK)
    for x in range(0, WIDTH, BLOCK):
        pygame.draw.line(screen, (20, 20, 20), (x, 0), (x, HEIGHT))
    for y in range(0, HEIGHT, BLOCK):
        pygame.draw.line(screen, (20, 20, 20), (0, y), (WIDTH, y))

    draw_snake(screen, snake)
    draw_food(screen, food)
    draw_score(screen, score)

    # Save key frames
    if frame in [0, 60, 130, 200, 260]:
        fname = f"task1_frame{frame}.png"
        pygame.image.save(screen, fname)
        screenshots.append(fname)

# Game Over screen
draw_gameover(screen, score)
pygame.image.save(screen, "task1_gameover.png")
screenshots.append("task1_gameover.png")

print(f"✅ Done! Final Score: {score}")
```

 Lab Task 2 — Advanced Snake Game

```python
import pygame
import random
from IPython.display import Image as IPImage, display

screen = pygame.display.set_mode((600, 400))

# Colors
BLACK  = (0,   0,   0)
WHITE  = (255, 255, 255)
GREEN  = (0,   200, 0)
DGREEN = (0,   150, 0)
RED    = (213, 50,  80)
YELLOW = (255, 255, 0)
BROWN  = (139, 69,  19)

WIDTH, HEIGHT = 600, 400
BLOCK = 20

font       = pygame.font.SysFont("arial", 24, bold=True)
small_font = pygame.font.SysFont("arial", 18)

# ── Functions ──────────────────────────────────────────
def draw_snake(surface, snake_list):
    for i, block in enumerate(snake_list):
        color = DGREEN if i == 0 else GREEN
        pygame.draw.rect(surface, color,
                        [block[0], block[1], BLOCK-2, BLOCK-2], border_radius=4)

def draw_food(surface, food):
    pygame.draw.circle(surface, RED,
                      (food[0] + BLOCK//2, food[1] + BLOCK//2), BLOCK//2)

def draw_bonus_food(surface, bonus):
    pygame.draw.circle(surface, YELLOW,
                      (bonus[0] + BLOCK//2, bonus[1] + BLOCK//2), BLOCK//2)
    label = small_font.render("x2", True, BLACK)
    surface.blit(label, (bonus[0]-2, bonus[1]-2))

def draw_walls(surface, walls):
    for wall in walls:
        pygame.draw.rect(surface, BROWN, wall)

def draw_score(surface, score, speed):
    text = font.render(f"Score: {score}  Speed: {speed}", True, WHITE)
    surface.blit(text, (8, 8))

def draw_gameover(surface, score):
    overlay = pygame.Surface((WIDTH, HEIGHT))
    overlay.set_alpha(180)
    overlay.fill(BLACK)
    surface.blit(overlay, (0, 0))
    go_text = font.render("GAME OVER", True, RED)
    sc_text = small_font.render(f"Final Score: {score}", True, WHITE)
    f1_text = small_font.render("Score >100: Speed x2 activated!", True, YELLOW)
    f2_text = small_font.render("Score >150: Bonus fruit appeared!", True, YELLOW)
    surface.blit(go_text, (WIDTH//2 - go_text.get_width()//2, HEIGHT//2 - 60))
    surface.blit(sc_text, (WIDTH//2 - sc_text.get_width()//2, HEIGHT//2 - 10))
    surface.blit(f1_text, (WIDTH//2 - f1_text.get_width()//2, HEIGHT//2 + 30))
    surface.blit(f2_text, (WIDTH//2 - f2_text.get_width()//2, HEIGHT//2 + 60))

# ── Walls ──────────────────────────────────────────────
walls = [
    pygame.Rect(200, 100, 20, 100),   # wall 1
    pygame.Rect(380, 200, 20, 100),   # wall 2
]

# ── Game State ─────────────────────────────────────────
snake       = [[200, 200], [180, 200], [160, 200]]
direction   = [BLOCK, 0]
food        = [random.randrange(1, (WIDTH//BLOCK)-1)  * BLOCK,
               random.randrange(1, (HEIGHT//BLOCK)-1) * BLOCK]
bonus_food  = None
score       = 0
speed       = 10
game_over   = False
screenshots = []

# ── Simulate 400 frames ────────────────────────────────
for frame in range(400):
    if game_over:
        break

    # Direction pattern
    if   frame < 80:  direction = [BLOCK,  0]
    elif frame < 140: direction = [0,  BLOCK]
    elif frame < 220: direction = [-BLOCK, 0]
    elif frame < 280: direction = [0, -BLOCK]
    elif frame < 340: direction = [BLOCK,  0]
    elif frame < 400: direction = [0,  BLOCK]

    new_head = [snake[0][0] + direction[0],
                snake[0][1] + direction[1]]
    snake.insert(0, new_head)

    # ── Feature 1: Score > 100 speed doubles ───────────
    if score > 100:
        speed = 20
    else:
        speed = 10

    # Normal food check
    if snake[0][0] == food[0] and snake[0][1] == food[1]:
        score += 10
        food = [random.randrange(1, (WIDTH//BLOCK)-1) * BLOCK,
                random.randrange(1, (HEIGHT//BLOCK)-1) * BLOCK]
    else:
        snake.pop()

    # ── Feature 2: Score > 150 bonus yellow fruit ──────
    if score > 150 and bonus_food is None:
        bonus_food = [random.randrange(1, (WIDTH//BLOCK)-1) * BLOCK,
                      random.randrange(1, (HEIGHT//BLOCK)-1) * BLOCK]

    if bonus_food:
        if snake[0][0] == bonus_food[0] and snake[0][1] == bonus_food[1]:
            score += 20   # double marks
            bonus_food = [random.randrange(1, (WIDTH//BLOCK)-1) * BLOCK,
                          random.randrange(1, (HEIGHT//BLOCK)-1) * BLOCK]

    # Wall collision
    for wall in walls:
        if wall.collidepoint(snake[0][0], snake[0][1]):
            game_over = True

    # Boundary collision
    if (snake[0][0] < 0 or snake[0][0] >= WIDTH or
        snake[0][1] < 0 or snake[0][1] >= HEIGHT):
        game_over = True

    # Self collision
    if snake[0] in snake[1:]:
        game_over = True

    # ── Draw ───────────────────────────────────────────
    screen.fill(BLACK)

    # Grid
    for x in range(0, WIDTH, BLOCK):
        pygame.draw.line(screen, (20, 20, 20), (x, 0), (x, HEIGHT))
    for y in range(0, HEIGHT, BLOCK):
        pygame.draw.line(screen, (20, 20, 20), (0, y), (WIDTH, y))

    draw_walls(screen, walls)
    draw_snake(screen, snake)
    draw_food(screen, food)

    if bonus_food:
        draw_bonus_food(screen, bonus_food)

    draw_score(screen, score, speed)

    # Save key frames
    if frame in [0, 80, 160, 240, 320]:
        fname = f"task2_frame{frame}.png"
        pygame.image.save(screen, fname)
        screenshots.append(fname)

# Game Over screen
draw_gameover(screen, score)
pygame.image.save(screen, "task2_gameover.png")
screenshots.append("task2_gameover.png")

print(f"✅ Done! Final Score: {score}  Speed: {speed}")
```

**Uniform Cost Search (UCS) Implementation**

Uniform Cost Search works by always expanding the node with the lowest cumulative path cost.

**Steps:**

Use a priority queue (min-heap) to store nodes based on cost.
Start from the initial node with cost = 0.
Repeatedly:
Remove the node with the smallest cost.
If it is the goal, return the path and cost.
Otherwise, expand its neighbors and add them to the queue.
Keep track of visited nodes to avoid reprocessing.

```python
import heapq

def ucs(graph, start, goal):
    # Priority queue: (cost, node, path)
    pq = [(0, start, [])]
    visited = set()

    while pq:
        cost, node, path = heapq.heappop(pq)

        if node in visited:
            continue

        visited.add(node)
        path = path + [node]

        # Goal check
        if node == goal:
            return cost, path

        # Expand neighbors
        for neighbor, weight in graph.get(node, []):
            if neighbor not in visited:
                heapq.heappush(pq, (cost + weight, neighbor, path))

    return float("inf"), []

# Graph from your diagram
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 8), ('E', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [],
    'E': [],
    'G': []
}

# Run UCS
cost, path = ucs(graph, 'S', 'G')

print("Shortest path:", path)
print("Total cost:", cost)
```

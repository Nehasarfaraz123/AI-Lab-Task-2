**Implementation of A* Algorithm in Python**

**Description:**
We created an unweighted graph.
We assigned heuristic values to each node.
We applied A* algorithm using f(n) = g(n) + h(n).
The algorithm finds the shortest path from A to H.

```python
import heapq

# Graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': ['H'],
    'F': [],
    'G': ['H'],
    'H': []
}

# Heuristic values
heuristic = {
    'A': 7,
    'B': 6,
    'C': 4,
    'D': 7,
    'E': 2,
    'F': 3,
    'G': 1,
    'H': 0
}

def a_star(start, goal):
    open_list = []
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))

    visited = set()

    while open_list:
        f, g, node, path = heapq.heappop(open_list)

        if node == goal:
            return path

        if node in visited:
            continue

        visited.add(node)

        for neighbor in graph[node]:
            new_g = g + 1  # unweighted graph cost = 1
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))

    return None


# Run A*
start = 'A'
goal = 'H'

result = a_star(start, goal)

print("Shortest Path using A* Algorithm:")
print(result)
```

A* Algorithm on Weighted Grap


Introduction

In this task, we implement the A search algorithm* on a weighted graph.
A* is an intelligent search algorithm used to find the shortest path between a start node and a goal node.

It uses two values:

g(n) = actual cost from start node
h(n) = heuristic (estimated cost to goal)
**f(n) = g(n) + h(n)**

A* always selects the path with the lowest f(n) value.

**Graph Representation (Weighted Graph)**

We represent the graph using an adjacency list, where each node contains its neighbors and edge weights.

**Heuristic Values**
Heuristic values estimate the distance from each node to the goal node G.

```python
# GRAPH WITH WEIGHTS
graph = {
    'S': {'A': 5, 'B': 2, 'C': 4},
    'A': {'D': 9, 'E': 4},
    'B': {'G': 6},
    'C': {'F': 2},
    'D': {'H': 7},
    'E': {'G': 6},
    'F': {'G': 1},
    'G': {},
    'H': {}
}

# HEURISTIC VALUES (estimated cost to goal G)
heuristic = {
    'S': 8,
    'A': 8,
    'B': 4,
    'C': 3,
    'D': 999,
    'E': 5,
    'F': 1,
    'G': 0,
    'H': 999
}
```

A* Algorithm Implementation

We use a priority queue (heap) to always expand the best node first.

```python
import heapq

def a_star(graph, heuristic, start, goal):
    # priority queue (f, g, node, path)
    open_list = []
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))

    visited = set()

    while open_list:
        f, g, current, path = heapq.heappop(open_list)

        # goal check
        if current == goal:
            return path, g

        visited.add(current)

        # explore neighbors
        for neighbor, cost in graph[current].items():
            if neighbor in visited:
                continue

            new_g = g + cost
            new_f = new_g + heuristic[neighbor]

            heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))

    return None, float("inf")
```

**Running the Algorithm**

Now we define the start and goal node and run the algorithm.

```python
start_node = 'S'
goal_node = 'G'

path, cost = a_star(graph, heuristic, start_node, goal_node)

print("Shortest Path:", path)
print("Total Cost:", cost)
```
